##encoding data

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
df['class']=(df['class']=='e').astype(int)#changing the expected output to 0 or 1
```

```python
ohe=OneHotEncoder(sparse_output=False)  #sparse_output=False le dataset ma convert garxha jasto lagyo
```

```python
ode=OrdinalEncoder()
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct=make_column_transformer(
    (ohe,nominal_data),#nominal_data ko sata list pathaunu paryo bhaney (ode,['cap-shape','cap-surface','cap-color','odor','gill-attachment','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-type','spore-print-color','habitat']
    (ode,ordinal_data),
    ('passthrough',['class']),
    remainder='drop'



)
```

```python
ct.set_output(transform="pandas")  #to change into dataframe
```

```python
df_transform=ct.fit_transform(df) ##data lai transform matra gareko ho yo
df_transform
```

```python
df_transform.columns[-1]
```

```python
#train valid and test dataset ma split garney using np.split
# .sample()le random values rakhxha
# train lai 60%
# valid lai 0.5 to 0.8 ie 20%
#test lai 20%

# train,valid,test=np.split(df_transform.sample(frac=1),[int(0.6*len(df_transform)),int(0.8*len(df_transform))])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df_transform.drop('passthrough__class',axis=1),df_transform['passthrough__class'],test_size=0.3,random_state=20)
```

##Feature Selection

```python
plt.figure(figsize=(12,10))
cor=X_train.corr()
sns.heatmap(cor,annot=True)
plt.show()
```

```python
def correlation(data,threshold=0.85):
  corr_set=set()
  corr_matrix=data.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if (corr_matrix.iloc[i,j])>threshold:
        corr_set.add(corr_matrix.columns[i])
  return corr_set
```

```python
corr_feature=correlation(X_train)
print(len(corr_feature))
```

```python
corr_feature
```

```python
X_train.drop(corr_feature,axis=1)
X_test.drop(corr_feature,axis=1)
```

```python
import numpy as np
import pandas as pd
import joblib
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer

# Load the trained model and transformer
model = joblib.load("mushroom_model.pkl")
encoder = joblib.load("transformer.pkl")

# User input
user_input = {
    'cap-shape': 'x', 'cap-surface': 's', 'cap-color': 'n', 'bruises': 't',
    'odor': 'p', 'gill-attachment': 'f', 'gill-spacing': 'c', 'gill-size': 'b',
    'gill-color': 'k', 'stalk-shape': 'e', 'stalk-root': 'c',
    'stalk-surface-above-ring': 's', 'stalk-surface-below-ring': 's',
    'stalk-color-above-ring': 'w', 'stalk-color-below-ring': 'w',
    'veil-type': 'p', 'veil-color': 'w', 'ring-number': 'o', 'ring-type': 'p',
    'spore-print-color': 'k', 'population': 's', 'habitat': 'u'
}

# Preprocess input
df = pd.DataFrame([user_input])

# Add a dummy 'class' column with a temporary value, it'll be dropped later
df['class'] = 0  # You can use any temporary value here

df_transformed = encoder.transform(df)

# Drop the temporary 'class' column (and any other unnecessary ones created during transformation)
# Keeping only the columns that were used during training
df_transformed = df_transformed[model.feature_names_in_]

# Predict
prediction = model.predict(df_transformed)

# Output result
if prediction[0] == 1:
    print("The mushroom is edible!")
else:
    print("The mushroom is poisonous!")
```
